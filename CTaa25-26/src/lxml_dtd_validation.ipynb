```python
from lxml import etree
```

```python
FOLDER_PATH="/content/drive/MyDrive/Colab/uni/"
XML_PATH="data/tei-doc.xml"
```

```python
parser = etree.XMLParser(dtd_validation=True, no_network=False)
```

```python
try:
    with open(FOLDER_PATH+XML_PATH, 'rb') as f:
            doc = etree.parse(f, parser)
            print(f"[OK] {FOLDER_PATH+XML_PATH} Ã¨ valido.")

except etree.XMLSyntaxError as e:
     # Questo errore cattura sia XML malformato che errori di validazione DTD
     print(f"[FALLITO] {FOLDER_PATH+XML_PATH}")
     # Recuperiamo il log degli errori dal parser per dettagli precisi
     for error in e.error_log:
        print(f"  - Riga {error.line}, Colonna {error.column}: {error.message}")

except OSError as e:
        print(f"[ERRORE SISTEMA] Non riesco a trovare la DTD o il file: {e}")
        print(f"  Suggerimento: Controlla che il percorso indicato nel DOCTYPE sia corretto rispetto alla posizione del file XML.")
```
